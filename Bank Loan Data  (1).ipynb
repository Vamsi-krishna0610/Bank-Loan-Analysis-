### Monthly Trends for Issue date for total amount

```python
monthly_Funds = (
    df.sort_values('issue_date')
      .assign(month_name = lambda x : x ['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort = False)['loan_amount']
      .sum()
      .div(1000000)
      .reset_index(name = 'Loan_Amount_Millions')
)

plt.figure(figsize = (10,5))
plt.fill_between(monthly_Funds['month_name'],monthly_Funds['Loan_Amount_Millions'], color = 'skyblue', alpha = 0.7)
plt.plot(monthly_Funds['month_name'],monthly_Funds['Loan_Amount_Millions'], color = 'Black', linewidth = 3)

for i,row in monthly_Funds.iterrows():
   plt.text(i, row['Loan_Amount_Millions'] + 0.1 , f"{row['Loan_Amount_Millions']:.2f}",
             ha = 'center', va= 'bottom', fontsize = 9, rotation = 0,color = 'black')
    
plt.title ('Total Funds Amount by Monthly', fontsize = 14)
plt.xlabel('Month')
plt.ylabel('Funds Amount (Millions)')
plt.xticks(ticks = range(len(monthly_Funds)), labels = monthly_Funds['month_name'], rotation = 45 )
plt.grid(True, linestyle ='--', alpha = 0.6 )
plt.tight_layout()
plt.show()
```

### Monthly Trends for Issue date by total amount Received

```python
Received_monthly = (
    df.sort_values('issue_date')
      .assign(month_name = lambda x : x ['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort = False)['total_payment']
      .sum()
      .div(1000000)
      .reset_index(name = 'Amount_Received_Millions')
)

plt.figure(figsize = (10,5))
plt.fill_between(Received_monthly['month_name'],Received_monthly['Amount_Received_Millions'], color = 'Yellow', alpha = 0.5)
plt.plot(Received_monthly['month_name'],Received_monthly['Amount_Received_Millions'], color = 'Black', linewidth = 3.5)

for i,row in Received_monthly.iterrows():
   plt.text(i, row['Amount_Received_Millions'] + 0.1 , f"{row['Amount_Received_Millions']:.2f}",
             ha = 'center', va= 'bottom', fontsize = 9.2, rotation = 0,color = 'Blue')
    
plt.title ('Total Funds Recevied by Monthly', fontsize = 15)
plt.xlabel('Month')
plt.ylabel('Recevied Amount (Millions)')
plt.xticks(ticks = range(len(Received_monthly)), labels = Received_monthly['month_name'], rotation = 45 )
plt.grid(True, linestyle ='--', alpha = 0.6 )
plt.tight_layout()
plt.show()
```

### Monthly Trends for Total Loan Applications 

```python
Applications_monthly = (
    df.sort_values('issue_date')
      .assign(month_name = lambda x : x ['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort = False)['id']
      .count()
      .reset_index(name = 'Loan_Applications_Count')
)

plt.figure(figsize = (10,5))
plt.fill_between(Applications_monthly['month_name'],Applications_monthly['Loan_Applications_Count'], color = 'Maroon', alpha = 0.3)
plt.plot(Applications_monthly['month_name'],Applications_monthly['Loan_Applications_Count'], color = 'Black', linewidth = 2)

for i,row in Applications_monthly.iterrows():
   plt.text(i, row['Loan_Applications_Count'] + 0.1 , f"{row['Loan_Applications_Count']:.2f}",
             ha = 'center', va= 'bottom', fontsize = 9.2, rotation = 0,color = 'black')
    
plt.title ('Total Loan Applications by Monthly', fontsize = 15)
plt.xlabel('Month')
plt.ylabel(' Loan Applications')
plt.xticks(ticks = range(len(Applications_monthly)), labels = Applications_monthly['month_name'], rotation = 45 )
plt.grid(True, linestyle ='--', alpha = 0.6 )
plt.tight_layout()
plt.show()
```

### Regional Analysis for Funding The Toatal Funding  

```python
state_funds = df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=True)
state_funds_Thousands = state_funds / 1000

plt.figure(figsize=(12, 8))  # wider and shorter â€” fits horizontal bars better
bars = plt.barh(state_funds_Thousands.index, state_funds_Thousands.values, color='red')

# Add labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 20,                     # a little more spacing
             bar.get_y() + bar.get_height()/2,
             f'{width:,.1f}k',
             va='center', fontsize=9)

# Titles and labels
plt.title('Total Funded Amount by State', fontsize=14, pad=15)
plt.xlabel('Funded Amount (in Thousands)', fontsize=12)
plt.ylabel('State', fontsize=12)

plt.tight_layout(pad=2.0)  # add a bit more padding
plt.show()
```
